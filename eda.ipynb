```
# Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```
# Loading the dataset from the given link
url = "https://raw.githubusercontent.com/midhunmohank/INFO6105/main/advertising.csv"
df = pd.read_csv(url)

```

```
# Exploring the first five rows of the dataset
print(df.head())

```

```
# Checking for null values in the dataset
print(df.isnull().sum())

```

```
# Summary statistics of the dataset
print(df.describe())

```

```
# Distribution of the target variable 'Sales'
sns.displot(df['Sales'])
plt.title('Distribution of Sales')
plt.show()

```

```
# Pairplot to visualize the relationship between the features and target variable
sns.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=4, aspect=1, kind='scatter')
plt.show()

```

```
# Correlation matrix to examine the correlation between the features
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()

```

```
# Boxplots to examine the distribution and outliers of the features
sns.boxplot(x=df['TV'])
plt.title('TV')
plt.show()

sns.boxplot(x=df['Radio'])
plt.title('Radio')
plt.show()

sns.boxplot(x=df['Newspaper'])
plt.title('Newspaper')
plt.show()
```
